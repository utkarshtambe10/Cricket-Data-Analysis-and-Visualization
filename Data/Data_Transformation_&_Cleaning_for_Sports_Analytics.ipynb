```python
#important libraries
import pandas as pd
import json
```

Match Summary

```python
with open('/content/drive/MyDrive/Cricket_Dataset/t20MatchResults.json') as c:
  data = json.load(c)

df_match = pd.DataFrame(data[0]['matchSummary'])
df_match.head()
```

```python
df_match.shape
```

```python
#using scorecard as match_id to link with other tables
df_match.rename({'scorecard': 'match_id'}, axis = 1, inplace = True)
df_match.head()
```

```python
#creating match_name column for better connecting with other tables
match_ids_dict = {}

for index, row in df_match.iterrows():
  key1 = row["team1"] + ' Vs ' + row["team2"]
  key2 = row["team2"] + ' Vs ' + row["team1"]

  match_ids_dict[key1] = row["match_id"]
  match_ids_dict[key2] = row["match_id"]
match_ids_dict
```

```python
df_match.to_csv('/content/drive/MyDrive/Cricket_Dataset/t20MatchResults.csv', index=False)
```

Batting Summary

```python
with open('/content/drive/MyDrive/Cricket_Dataset/t20BattingSummary.json') as c:
  data = json.load(c)

  all_records = []
  for record in data:
    all_records.extend(record['battingSummary'])

df_batting = pd.DataFrame(all_records)
df_batting.head(11)
```

```python
#adding a new column of dismissal decision(out <-> not-out)
df_batting["out/not_out"] = df_batting.dismissal.apply(lambda x: "out" if len(x) > 0 else "not_out")
df_batting.head(11)
```

```python
#dropping the dismissal column
df_batting.drop(columns=["dismissal"], inplace=True)
df_batting.head(11)
```

```python
#removing the special characters from names of the players
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('â€', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('\xa0', ''))
df_batting['batsmanName'] = df_batting['batsmanName'].apply(lambda x: x.replace('†', ''))
df_batting.head(11)
```

```python
df_batting["match_id"] = df_batting["match"].map(match_ids_dict)
df_batting.head()
```

```python
df_batting.shape
```

```python
#saving the above results as CSV file
df_batting.to_csv('/content/drive/MyDrive/Cricket_Dataset/t20BattingSummaryWithMatch.csv', index=False)
```
